### Introduction : 

DBSCAN : Density-Based Spatial Clustering of Applications with Noise

La plupart des techniques traditionnelles de regroupement, telles que les k-means, le regroupement hiérarchique et fuzzy,
peut être utilisé pour regrouper des données de manière non supervisée.

Cependant, lorsqu'il est appliqué à des tâches avec des groupes de formes arbitraires, ou des groupes à l'intérieur de groupes,
les techniques traditionnelles pourraient ne pas être en mesure d'obtenir de bons résultats.
En d'autres termes, les éléments d'un même groupe peuvent ne pas être suffisamment similaires - ou les performances
peut être pauvre.

En outre, si les algorithmes basés sur le partitionnement, tels que K-Means, peuvent être faciles à comprendre
et l'implémenter dans la pratique, l'algorithme n'a aucune notion de valeurs aberrantes.
C'est-à-dire que tous les points sont affectés à un groupe, même s'ils n'appartiennent à aucun.

Dans le domaine de la détection des anomalies, cela pose des problèmes car les points anormaux seront attribués au même groupe que les points de données "normaux". Les points anormaux tirent le centroïde de l'amas vers eux, ce qui rend plus difficile leur classification en tant que points anormaux.
En revanche, le regroupement basé sur la densité permet de localiser les régions à forte densité qui sont séparées les uns des autres par les régions de faible densité. La densité, dans ce contexte, est définie comme la nombre de points dans un rayon déterminé. Un type spécifique et très populaire de regroupement est DBSCAN.

DBSCAN signifie Density-Based Spatial Clustering of Applications with Noise.
Cette technique est l'un des algorithmes de clustering les plus courants, qui fonctionne sur la base de la densité des objet. DBSCAN travaille sur l'idée que si un objet particulier appartient à un groupe, il doit être proche de nombreux autres points de ce groupe.
Il fonctionne sur la base de 2 paramètres : Rayon et Points minimums.
R détermine un rayon spécifié qui, s'il comprend suffisamment de points, est appelé c'est une "zone dense". M détermine le nombre minimum de points de données que nous voulons dans un quartier pour définir un groupe.

### Avantages de DBSCAN : 

DBSCAN peut trouver des clusters de forme arbitraire. Il peut même trouver un groupe complètement entourée par un groupe différent. DBSCAN a une notion de bruit, et est robuste aux valeurs aberrantes. En outre, le DBSCAN rend très pratique
pour être utilisé dans de nombreux problèmes réellement mondiaux, car il n'est pas nécessaire de préciser le nombre de groupes, comme K dans k-Means.
